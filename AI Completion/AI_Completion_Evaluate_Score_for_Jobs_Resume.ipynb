### **Extract text from document**
---

````python
pip install pypdf
````

````python
from pypdf import PdfReader

reader = PdfReader("/content/resume.pdf")

page_texts = []
for page in reader.pages:
    text = page.extract_text()
    page_texts.append(text if text else "")

resume_text = "\n-----\n".join(page_texts)

# example
print(f"จำนวนหน้า: {len(page_texts)}")
print("ข้อความรวมทั้งหมด:")
print()
print(resume_text)
````

### **Requests**
---

Agent Name: **Ex-HR Specialist**

````python
import requests


position_name = "Data Engineer"

bearer_token = "API_KEY_HERE"  # Replace with your actual API key
url = "https://genai.softnix.ai/external/api/completion-messages"
payload = {
  "inputs": {"position":position_name,"resume_text":resume_text},
  "citation": True,
  "response_mode": "blocking"
}

headers = {
  "Content-Type": "application/json",
  "Authorization": f"Bearer {bearer_token}"
}
response = requests.post(url, json=payload, headers=headers, verify=False)
````
